# quiz2

소방차가 앰뷸런스보다 8분 이내에 도착하는 비율이 높다.

```python
import pandas as pd
df2 = pd.read_excel("/content/drive/MyDrive/통계학개론_오태훈/퀴즈/quizdata2206511.xlsx",
                     sheet_name='ambulance')
df2.info()
```

```python
df_fire = df2[df2['Fire'] < 8].shape[0]
```

```python
df_ambulance = df2[df2['Ambulance'] < 8].shape[0]
```

```python
n = df2.shape[0]
```

```python
from statsmodels.stats.proportion import proportions_ztest
stat, pval = proportions_ztest(count = (df_fire, df_ambulance),nobs = (n,n),value = 0, alternative = 'larger')
```

```python
if pval >= 0.05:
  conclude = '귀무가설 채택';
else:
  conclude = '귀무가설 기각';
print('1집단 표본비율 = %4.2f, 2집단 표본비율 = %4.2f, 검정통계량 = %3.3f, 유의확률 = %.3f, 결론: %s' %(df_fire/n, df_ambulance/n, stat, pval,conclude))
# 유의확률이 0.068 이는 유의수준인 0.05보다 크므로 귀무가설을 채택, 소방차 도착비율과 엠뷸런스 도착비율이 같다. 심장박동기를 장착하면 안된다.
```

```python
#귀무가설 : 소방차와 앰뷸런스의 8분이내의 도착 비율이 같다.
#대립가설 : 소방차가 앰뷸런스보다 8분 이내의 도착 비율이 높다.
x=[(df2['Fire']<=8).sum(),(df2['Ambulance']<=8).sum()]
n=[df2['Fire'].count(),df2['Ambulance'].count()]
from statsmodels.stats.proportion import proportions_ztest
proportions_ztest(x,n,alternative='larger')
#유의확률이 0.068 이는 유의수준인 0.05보다 크므로 귀무가설을 채택.
# 소방차 도착비율과 앰뷸런스 도착비율이 같다. 따라서 심장박동기를 장착하면 안된다.
```

# 타이타닉

```python
titanic = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/carData/TitanicSurvival.csv')
titanic.info()
```

```python
titanic['sex'].unique()
```

```python
# 문제 : 여자가 남자보다 생존율이 높은가
result = pd.crosstab(titanic["sex"],titanic["survived"],margins=True)
result.iloc[0,0]
```

```python
result
```

```python
#문제:여자가 남자보다 생존율이 높은가.
#귀무가설: 여자생존율과 남자생존율은 동일하다.
#대립가설: 여자생존율이 남자생존율보다 높다.
x=[result.iloc[0,1],result.iloc[1,1]]
n=[result.iloc[0,2],result.iloc[1,2]]
from statsmodels.stats.proportion import proportions_ztest
proportions_ztest(x,n,alternative='larger')
#유의확률 0.001 < 유의수준보다 작으므로 귀무가설을 기각. 따라서 여자생존율이 남자생존율보다 높다.
```
